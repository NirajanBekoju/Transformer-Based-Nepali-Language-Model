# Process and save the entire text

```python

limit = 100300
t = ''
progress_bar = tqdm(enumerate(text.split('\n')[:limit]),total = limit, desc = 'processing', ncols=80)
for i,u in progress_bar:    
    
    t+= '-'.join(models[0].viterbi_segment(u)[0])
    t+='\n'
    progress_bar.set_postfix({"iteration_num": i}, refresh=True)

```

```python
with open('data/preprocessed_morph.txt', 'w', encoding='utf-8') as f:
    f.write(t)
```
